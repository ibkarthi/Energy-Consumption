```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import geopandas as gpd
```

```python
world_df = gpd.read_file('World_Countries.shp')
```

```python
world_df
```

```python
plt.figure(figsize = (10,6), dpi = 150)
world_df.plot(figsize = (10,6));
```

```python
world_df.columns = ['country', 'geometry']
```

```python
world_df.head()
```

```python
r_energy = df.merge(world_df, on = 'country')
```

```python
renew_gdf = gpd.GeoDataFrame(r_energy, geometry = r_energy['geometry'])
```

```python
renew_gdf.reset_index(drop = True, inplace = True)
```

```python
renew_gdf.head()
```

```python
plt.figure(figsize = (25,8), dpi = 150)
renew_gdf.plot(column = 'wind_consumption', cmap = 'Pastel1', figsize = (25,8), legend = True)
plt.title('Global Wind Energy Consumption in Terra-Watt Hour', fontsize = 25);
```

```python
plt.figure(figsize = (25,8), dpi = 150)
renew_gdf.plot(column = 'solar_consumption', 
               cmap = 'YlOrRd', 
               figsize = (25,8), 
               legend = True)
plt.title('Global Solar Energy Consumption in Terra-Watt Hour', fontsize = 25);
```

```python
plt.figure(figsize = (25,10), dpi = 150)
renew_gdf.plot(column = 'hydro_consumption', 
               cmap = 'Blues', 
               figsize = (25,10), 
               legend = True)
plt.title('Global Hydro Energy Consumption in Terra-Watt Hour', fontsize = 25);
```

```python
renew_gdf.columns
```

```python

plt.figure(figsize = (25,8), dpi = 150)
renew_gdf.plot(column = 'other_renewable_consumption', 
               cmap = 'Greens', 
               figsize = (25,8),
               edgecolor='gainsboro',
               legend = True)
plt.title('Global Other Renewable Energy Consumption in Terra-Watt Hour', fontsize = 25);
```

```python
plt.figure(figsize = (25,8), dpi = 150)
renew_gdf.plot(column = 'renewables_consumption', 
               cmap = 'YlGn', 
               figsize = (25,8),
               edgecolor='gainsboro',
               legend = True)
plt.title('Global Overall Renewable Energy Consumption in Terra-Watt Hour', fontsize = 25);
```
